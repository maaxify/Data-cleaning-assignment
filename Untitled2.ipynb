## Renaming the column headers and after that putting them in the right datatype

### The Date datatype is not correct so, let me correct it.

```python
data.Date
```

```python
data.Date=pd.to_datetime(data.Date).dt.strftime('%d/%m/%y')
data.Date
```

## The payment Type 

```python
data['PaymentType'].value_counts()
```

```python
data['PaymentType']=data['PaymentType'].replace(['Debit','debit card'],'Debit card')
data['PaymentType']=data['PaymentType'].replace(['Credit Card','Credit'],'Credit card')
```

```python
data['PaymentType'].value_counts()
```

## The Negative values

```python
data=data[(data['Quantity']>0) & (data['Customer_Age']>0)]
data
```

## Finding and correcting Typographical errors
